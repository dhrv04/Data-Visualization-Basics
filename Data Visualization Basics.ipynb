### Pie Plot

```python
labels = ['Frogs', 'Cats', 'Dogs', 'Lions']
sizes = [15, 30, 45, 80]

fig, ax = plt.subplots(figsize=(4,4))
explode = (0, 0.1, 0, 0)
ax.pie(sizes, labels=labels, autopct='%1.1f%%', radius=1.5, shadow=True, startangle=60, explode=explode)
plt.show()
```

```python
# Code for changing font size and color of texts

labels = ['Frogs', 'Cats', 'Dogs', 'Lions']
sizes = [15, 30, 45, 80]

fig, ax = plt.subplots(figsize=(4,4))
explode = (0, 0.1, 0, 0)
patches, texts, autotexts = ax.pie(sizes, labels=labels, autopct='%1.1f%%', radius=1.5, shadow=True, startangle=60,
                                   explode=explode)

plt.setp(autotexts, size=12, color = 'black')
autotexts[1].set_color('white')
plt.show()
```

### Labelling Pie Chart

```python
fig, ax = plt.subplots(dpi=150)
recipe = ["375 g Flour", "75 g Sugar", "250 g Butter", "300 g Berries"]

data = [float(x.split(' ')[0]) for x in recipe]
ing = [x.split(' ')[-1] for x in recipe]

def func(pct, vals):
    absolute = int(pct/100.*np.sum(vals))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)

wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data), textprops=dict(color='w'))

ax.legend(wedges, ing, title='Ingredients', loc='upper right', bbox_to_anchor=(1,0,1,1))

plt.show()
```

###  Bar Chart on Polar Axis

```python
np.random.seed(0)
N = 30
theta = np.linspace(0.0, 2*np.pi, N, endpoint=False)
radii = 10*np.random.rand(N)
width = np.pi/4*np.random.rand(N)
colors = plt.cm.plasma(radii/10)

plt.figure(dpi=100)
ax = plt.subplot((111), projection='polar')
ax.bar(theta, radii, width=width, bottom=0, color=colors, alpha=0.7)
plt.show()
```

### Line Plot on Polar Axis

```python
r = np.arange(0, 2, 0.01)
theta = 2*np.pi*r

plt.figure(dpi=100)
ax = plt.subplot(111, projection='polar')
ax.plot(theta, r)
ax.set_rmax(2)
ax.set_rticks([0, 0.5, 1, 1.5])
ax.set_rlabel_position(-30)
ax.grid(True)

plt.show()
```
