```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
from plotly.offline import iplot
import plotly as py
import cufflinks as cf
```

```python
py.__version__
```

```python
py.offline.init_notebook_mode(connected=True)
cf.go_offline()
```

### Line Plot

```python
df = pd.DataFrame(np.random.randn(100, 3), columns=['A', 'B', 'C'])
```

```python
df.head()
```

```python
df = df.cumsum()
```

```python
df.head()
```

```python
df.iplot()
```

```python
df.iloc[10]
```

```python
# Compared to the normal plot

df.plot()
```

### Scatter Plot

```python
df.head()
```

```python
df.iplot(x='A', y='B', mode='markers', title='Scatter Plot', xTitle='X', yTitle='Y')
```

### Bar Plot

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.iplot(kind='bar', x='sex', y='survived', title='Titanic')
```

```python
df.iplot(kind='bar', barmode='stack', bargap=0.5)
```

```python
df.iplot(kind='barh', barmode='stack', bargap=0.5)
```

### Box Plot

```python
df.iplot(kind='box')
```

### Area Plot

```python
df.iplot(kind='area', fill=True)
```

### 3D Plots

```python
df.head()
```

```python
df.iplot(kind='surface', colorscale='rdylbu')
```

```python
cf.datagen.sinwave(10, 0.25).iplot(kind='surface')
```

### Spread Plot

```python
df.iplot(kind='spread')
```

```python
df.iplot(kind='hist', bins=25, barmode='overlay', bargap=0.25)
```

### Bubble Plot

```python
bb = cf.datagen.bubble3d(5,4, mode='stocks')
```

```python
bb.head()
```

```python
bb.iplot(kind = 'bubble3d', x = 'x', y= 'y', z = 'z', size = 'size')
```

### HeatMap

```python
df.corr()
```

```python
df.corr().iplot(kind='heatmap', colorscale='spectral')
```

```python
cf.datagen.heatmap(20, 20).iplot(kind = 'heatmap', colorscale='spectral')
```
